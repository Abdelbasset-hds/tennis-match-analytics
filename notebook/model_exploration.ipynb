```python
from ultralytics import YOLO
```

Tried YOLOv8x on tennis ball detection.
It doesn’t work that well — most frames are missed, only a few correct detections.
Looks like the pretrained model isn’t really suited for tennis balls, so custom training will probably be needed.

```python
model = YOLO('yolov8x')
results = model.predict('../data/tennis_match_lite.mp4',save = True)
print(results)
print("boxes")
for box in results[0].boxes :
    print(box)
```

```python
model = YOLO('../models/best.pt')
results = model.predict('../data/tennis_match_lite.mp4',save = True)
print(results)
print("boxes")
for box in results[0].boxes :
    print(box)
```
